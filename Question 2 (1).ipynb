```python
# Part 1

import pandas as pd
import numpy as np
df = pd.read_csv('./data/airbnb_hw.csv')
print(df.dtypes)      # now we know that price is an object
```

```python
selection = df[['Host Id','Price']]
print(selection)
print(selection.dtypes)
```

```python
nan_indices = selection.isna()  # i used this block of code to see if there were any NANs in the price data. 
print(nan_indices)             # looking like no
print('\n')
summ = nan_indices.sum()
print(summ)

for value in nan_indices: 
    if value==True:
        print("NAN in set")
    else: 
        continue
print('\n')
print('Completed Nan Check')
```

```python
# price_nan= df['Price'] = pd.to_numeric(df['Price'], errors='coerce').fillna(0).astype(int)
# print(price_nan)

comma_cols=[]
for col in selection.columns:
    if selection[col].dtype == 'object':
        if selection[col].str.contains(',').any():
            comma_cols.append(col)

# comma_cols is just price. Trying to change from an object to an integer

# choices=selection.loc[[0],['Price']]
# print(choices)

# comma_cols is only telling us that price has commas. so now we just need to get into that column specifically. 

new_price=[]


for price in selection['Price']:
    if ',' not in price:
        new_price.append(price)
    else: 
        newest_price = price.replace(',', '')
        new_price.append(newest_price)

df['New Price']=new_price

df['New Price'] = df['New Price'].astype(int)
print(df['New Price'])
```

## Choices explained

For this question, I isolated the price column, checked for NAs, and found commas. Then, I made a new column in the 
dataset called 'New Price' to make it easier to work with the data and not change the dataset directly. I then changed
the new "New Price" column into an integer so that we could do numerical analysis with it if we wanted to.
