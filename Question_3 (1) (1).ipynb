```python
## (1) 
# Loaded data with Pandas

import pandas as pd

df = pd.read_csv('./data/college_completion.csv')
print(df['grad_100_value'])
```

```python
## (2)
# Used head to examine the data and offer descriptions

df.head()
# print(df.shape)
# Dimensions: The dimensions of the data are 3798 by 63
# Observations: The number of observations is 3798
# Variables: There are 63 variables in the data set
```

```python
## (3)
# Cross tabulated control and level variables and gave description

cross_tab = pd.concat([df['level'],df['control']],axis=1)
cross_tab.tail()
df.loc[50:100, ['level', 'control']]
# some of the patterns that I see are the Private not-for-profit options seem to have the most 4-year levels,
# the public schools seem to have the highest amount of 2-year levels, and the Private for-profit seem to have more 4-year 
# level than 2-year. 
```

```python
## (4)
# grad_100_value histogram

import matplotlib.pyplot as plt

# histogram
df['grad_100_value'].plot.hist(bins=5, edgecolor='black')

plt.xlabel('% of Students Who Graduated on Time')
plt.ylabel('Frequency')
plt.title('On-Time Graduates Frequency')
```

```python
## (4)
# grad_100_value kernel density plot

df['grad_100_value'].plot.kde()

plt.xlabel('% On-Time Graduated Students')
plt.ylabel('Density')
plt.title('% On Time Graduated Students Kernel Density Plot')
```

```python
## (4)
# grad_100_value boxplot

import seaborn as sns

# boxplot
sns.boxplot(data=df['grad_100_value'])

plt.xlabel("College Completion Data")
plt.ylabel("% On-Time Graduating Students")
plt.title("% On-Time Graduating Students Box Plot")
```

```python
## (4)
# Statistical description of histogram, kernel density plot, and boxplot:

df['grad_100_value'].describe()

# The mean: 28.36
# The median: 22.5 
# Minimum: 0
# Maximum: 100
# Standard Deviation: 23.3
# All of the prior plots demonstrate a huge skew to the right: most institutions have an On-Time percent graduation rate
# of below 25 %. Institutions with a % On-Time graduation rate of higher than 97 are outliers according to these data. 
# However, the standard deviation from the mean is large and denotes that there is an obtuse variability in the 
# percentages of On-Time graduating students at institutions mentioned in these data. Also, 75% of institutions have 
# graduation rates that are under 45%, which indicates extremely poor On-Time graduation rates for most of the institutions
# mentioned in these data. 
```

```python
## (5)
# Kernel density plot of grad_100_value by control

sns.kdeplot(data=df, x='grad_100_value', hue='control', fill=True)

plt.xlabel('% On-Time Graduated Students')
plt.ylabel('Density')
plt.title('% On Time Graduated Students Kernel Density Plot')
```

```python
## (5)
# Kernel density plot of grad_100_value by level

sns.kdeplot(data=df, x='grad_100_value', hue='level', fill=True)

plt.xlabel('% On-Time Graduated Students')
plt.ylabel('Density')
plt.title('% On Time Graduated Students Kernel Density Plot')
```

```python
## (5) Describing what I see:

## According to the kernel density plots, it seems as though 4-year universities have a higher on-time graduation rate 
# percentage than 2-year univerisities, and private not-for-profit universities have the higher on-time graduation rate 
# percentage than both public and private for-profit universities. 
```

```python
## (5)
# Statistical descriptions of grad_100_value by control and level with best rated institution obseration

control_grouping = df.groupby('control')['grad_100_value'].describe()
level_grouping = df.groupby('level')['grad_100_value'].describe()

print(control_grouping)
print(level_grouping)

## Institutions that appear to have the best graduation rates: 
# Four year, private not-for-profit institutions appear to have the best graduation rates according to these data. 
```

```python
## (6)
# Grouped kernel density plot with new variable and best graduation rate institution prediction

df['levelXcontrol']=df['level']+', '+df['control']
sns.kdeplot(data=df, x="grad_100_value", hue="levelXcontrol", fill=True, common_norm=False)

plt.xlabel('% On-Time Graduated Students')
plt.ylabel('Density')
plt.title('% On Time Graduated Students Kernel Density Plot')

## According to this mixed kernel density plot, it seems as though 2 year private not-for profit institutions have the 
# highest density of graduating students, but the 2 year private for-rpofit and the 4 year private not-for-profit are
# close behind. This means that the 2 year programs at private not-for-profit and private for-profit have the best overall
# graduation rates according to these data. 
```

```python
## (7)
# kernel density plot of aid value

df['aid_value'].plot.kde()

plt.xlabel('Aid Value')
plt.ylabel('Density')
plt.title('Aid Value of Students')
```

```python
## (7)
# aid value grouped by control 

sns.kdeplot(data=df, x='aid_value', hue='control', fill=True)

plt.xlabel('Aid Value')
plt.ylabel('Density')
plt.title('Aid Value of Students by Institution Type')
```

```python
## (7)
# aid value grouped by level 

sns.kdeplot(data=df, x='aid_value', hue='level', fill=True)

plt.xlabel('Aid Value')
plt.ylabel('Density')
plt.title('Aid Value of Students by Program Length')
```

```python
## (7)
# Statistical descriptions of aid value by level and control and bi-modal-ish nature description

# The bimodal nature of the graphs is explained by private not-for-profit and 4-year institutions. Mostly because the
# main peak for private not-for-profit institutions lies around 15,000. 4 Year institutions also seem to have a small 
# bimodal peak around 15,000 as well, but it is not as pronounced as the private not-for-profit peak. 

control_aid_grouping = df.groupby('control')['grad_100_value'].describe()
level_aid_grouping = df.groupby('level')['grad_100_value'].describe()

print(control_aid_grouping)
print('\n---- Next Set ----\n')
print(level_aid_grouping)

# Statistical Description:
# private not-for-profit institutions and 4 year programs have the largest mean with 41.6 and 33.2 % respectively. However, 
# these level and control variable groups alos have a standard deviation of around 23 which is large considering. There is 
# also no public institution within these data that have a maximum % On-Time graduation rate of 100%. The 50% quartiles
# for 4 year and private not-for-profit institutions also have the largest values with 29.7 and 41 % respectively. This indicates
# that 4 year and private not-for-profit institutions demonstrate the highest % of On-Time graduating students according
# to this data set. 
```
